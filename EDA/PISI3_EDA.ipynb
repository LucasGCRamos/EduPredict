## analise de variaveis socioeconomicas


Nesta etapa foi analisado o comportamento das seguintes colunas:
* Mother's qualification
* Father’s qualification
* Mother’s occupation
* Father’s occupation

Essas colunas possuem muitos valores unicos, então foi necessario realizar uma filtragem para que a visualização em graficos ficasse mais limpa.

```python
columns = ["Mother's qualification","Father's qualification","Mother's occupation","Father's occupation"]
for column in columns:
  plot_data = dataset[column].copy()
  value_counts = plot_data.value_counts()
  plot_data = plot_data.replace(value_counts[value_counts < 20].index, 'Other')

  # Preparar os dados para o gráfico
  plot_data = plot_data.value_counts().reset_index()
  plot_data.columns = [column, 'Count']
  # Aplicando quebra de linha nos espaços
  plot_data[column] = plot_data[column].apply(
    lambda x: '<br>'.join(x.split(' '))
)

  # Criar o gráfico
  fig = px.bar(plot_data, x=column, y="Count",
              title=f"Distribuição de {column}")
  fig.update_layout(xaxis=dict(tickangle=0))
  fig.show()



```

* Educational special needs
* Debtor
* Tuition fees up to date
* Scholarship holder

```python
socioeconomic_variables = ["Debtor","Tuition fees up to date","Scholarship holder","Educational special needs"]
for column in socioeconomic_variables:
    fig = px.bar(dataset[column].value_counts(), y="count", title=f"Proporção de {column}")
    fig.update_layout(height=400, width=800, showlegend=True)
    fig.show()
```

## Variaveis sobre admissão na universidade
Nesta etapa, foi analisado o comportamento das seguintes colunas:
* Application mode
* Application order
* Course
* Daytime/evening attendance
* Previous qualification

```python
columns = ["Application mode","Previous qualification","Course","Application order"]
for column in columns:
  plot_data = dataset[column].copy()
  value_counts = plot_data.value_counts()
  plot_data = plot_data.replace(value_counts[value_counts < 100].index, 'Other')

  # Preparar os dados para o gráfico
  plot_data = plot_data.value_counts().reset_index()
  plot_data.columns = [column, 'Count']

  # Criar o gráfico
  fig = px.bar(plot_data, x=column, y="Count",
              title=f"Distribuição de {column}")
  fig.update_layout(height=400, width=800, showlegend=True)
  fig.show()
```

## Relação de cada variável com variável alvo
Nesta etapa, foi analisado a relação entre algumas variaveis demograficas e socioeconomicas com a nossa variavel target

```python
columns = ["Gender", "Daytime/evening attendance\t","Debtor", "Displaced","Scholarship holder" ]
for column in columns:
    fig = px.histogram(
        dataset,
        x=column,
        color='Target',
        barmode='group',
        title=f"Distribuição de {column} por Target")
    fig.update_layout(height=400, width=800, showlegend=True)
    fig.show()
```

## Heatmap da correlação entre variáveis de desempenho acadêmico

```python
selected_vars = [
    'Curricular units 1st sem (enrolled)', 'Curricular units 1st sem (approved)',
    'Curricular units 2nd sem (enrolled)', 'Curricular units 2nd sem (approved)',
    'Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)'
]

# Matriz de correlação
corr_matrix_selected = dataset[selected_vars].corr()


fig = px.imshow(
    corr_matrix_selected,
    color_continuous_scale="RdBu",
    zmin=-1, zmax=1,
    title="Heatmap das Variáveis Relacionadas ao Desempenho Acadêmico"
)

fig.show()
```
